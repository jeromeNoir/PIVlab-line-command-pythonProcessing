```python
import numpy as np
from scipy.io import loadmat
```

```python
def Data_multiple_frames(file_path: str, frame: int):
    # Load .mat
    mat = loadmat(file_path)

    # Frame index (1-based -> 0-based)
    frame_ind = frame - 1  

    # Coordinates
    x_all = mat['x']
    y_all = mat['y']
    X_original = x_all[frame_ind, 0]
    Y_original = y_all[frame_ind, 0]

    # Cell arrays: pick the frame cell
    U_filtered = mat['u_filtered'][frame_ind, 0]
    V_filtered = mat['v_filtered'][frame_ind, 0]
    U_org = mat['u_original'][frame_ind, 0] #velocity prior to validation
    V_org = mat['v_original'][frame_ind, 0] #velocity prior to validation

    # Flip axes
    X = X_original[:, ::-1]
    Y = Y_original[::-1, :]
    U = U_filtered[::-1, ::-1] #velocity after validation
    V = V_filtered[::-1, ::-1] #velocity after validation
    U_original = U_org[::-1, ::-1] #velocity prior to validation
    V_original = V_org[::-1, ::-1] #velocity prior to validation

    # Mask wherever original vectors are NaN
    mask = np.isnan(U_original) | np.isnan(V_original)
    # Ensure float dtype to hold NaNs
    U = U.astype(float, copy=False)
    V = V.astype(float, copy=False)
    X = X.astype(float, copy=False)
    Y = Y.astype(float, copy=False)
    # Apply mask elementwise
    U[mask] = np.nan
    V[mask] = np.nan
    X[mask] = np.nan
    Y[mask] = np.nan

    return X, Y, U, V, U_original, V_original, X_original, Y_original

def Data_single_frame(file_path: str):
    """Load data from a PIVlab .mat file that has only one frame."""
    mat = loadmat(file_path)
    X_original = mat['x'] 
    Y_original = mat['y']

    U_original = mat['u_original'] #velocity prior to validation
    V_original = mat['v_original'] #velocity prior to validation
    U = mat['u_filtered'] #velocity after validation
    V = mat['v_filtered'] #velocity after validation

    # Flip axes
    X = X_original[:, ::-1]
    Y = Y_original[::-1, :]
    U = U[::-1, ::-1] #velocity after validation
    V = V[::-1, ::-1] #velocity after validation
    U_original = U_original[::-1, ::-1] #velocity prior to validation
    V_original = V_original[::-1, ::-1] #velocity prior to validation
    # Mask wherever original vectors are NaN
    mask = np.isnan(U_original) | np.isnan(V_original)
    # Ensure float dtype to hold NaNs
    U = U.astype(float, copy=False)
    V = V.astype(float, copy=False)
    X = X.astype(float, copy=False)
    Y = Y.astype(float, copy=False)
    # Apply mask elementwise
    U[mask] = np.nan
    V[mask] = np.nan
    X[mask] = np.nan
    Y[mask] = np.nan

    return X, Y, U, V, U_original, V_original, X_original, Y_original
```
